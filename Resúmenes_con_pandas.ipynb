Utilizando documento csv

```python
import pandas as pd
personajes=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datos/characters.csv')
```

vamos a crear una nueva columna donde la altura de los personajes se encuentre en metros

```python
personajes['height_m']=personajes['height']/100
personajes.head()
personajes['mass'][15]
```

Ahora queremos obtener el indice de masa de los personajes de star wars del documento

```python
personajes['mass']=personajes['mass'].str.replace(',', '').astype(float)# valores con ,
personajes['imc']=personajes['mass']/(personajes['height_m'])**2
personajes.head()
```

Otra práctica de utilidad al crear columnas derivadas es el empleo de condicionales para el registro de valores en dependencia de su veracidad o falsedad.
Para que puedas realizar este proceso debes utilizar una notación especial, la cual puede resumirse de la siguiente manera:

$[valor\_verdadero$ **if** $condición$ **else** $valor\_falso$ **for** $valor$ **in** $columna]$

Ejemplo. Si quisieramos crear una columna "vacuna" donde pongamos "aplicada" en donde el género sea **female**; en caso contrario colocariamos **no** para todos los valores de "género" del dataframe. Sería

```python
personajes['vacuna']=["Aplicada" if gender == "female" else "No" for gender in personajes['gender']]
personajes.head()
```

# Resúmenes

Esta labor es fundamental en el proceso de ciencia de datos para **extraer conocimiento de la información que se procesa o lograr un mejor entendimiento de tus datos.**

**Solo tienes que poner especial cuidado de que los cálculos se efectúen sobre valores operables numéricamente**.

En una columna (función mean() en pandas)

```python
print("La altura promedio de los personajes de star wars es: ", personajes["height_m"].mean().round(2), 'metros')
```

Este tipo de funciones se aplican a varias columnas:

```python
print('Mediana de las alturas y peso de los personajes: ', personajes[["height_m","mass"]].median(),sep='\n')
```

También se pueden aplicar varias funciones a una o más columnas a la vez. Para esto se utiliza el método **agg()** sobre el dataframe.

```python
personajes.agg(
    {
        "height":['min','max','median','skew'],
        'mass': ['min','max','median','skew'],
    }
)
```

Estas funciones también pueden ser aplicadas a los resultados de un **agrupamiento**; es decir, **ejecutar alguna función sobre una columna en grupos determinados por el valor de otra columna**. En otras palabras, los registros de una columna se dividen en grupos, según el contenido de la segunda.

Por ejemplo, si quisieramos obtener el promedio de peso de los personajes de Star Wars pero agrupados por su género, entonces :

```python
personajes[['gender','mass']].groupby("gender").mean()
```
