# Filas

En el caso en que desees mantener todas las caracter√≠sticas o atributos (columnas) de un registro, pero no todos los registros, es decir, **seleccionar filas por condici√≥n**.
Para lograrlo, es necesario el uso de condicionales a algunos de los atributos o caracter√≠sticas de los registros

El filtrado por la condici√≥n especificada **incluir√° a aquellos registros referidos como filas que la cumplen y excluir√° a los que no lo hacen**. Debes colocar el condicional dentro de los corchetes y podr√°s utilizar los operadores relacionales $<,>,==,!=,>= ,<=$ que ya conoces.
Por ejemplo, podemos obtener todos los personajes de Star Wars que sean m√°s bajos de 100 cm como se muestra en el siguiente script:

```python
personas_bajas=starw[starw['height']<100]
print('Tama√±o del producto ', personas_bajas.shape)
personas_bajas
#mantiene todas las columnas pero solo hay 7 filas en el resultado
```

Tambi√©n se puede utilizar **condicionales compuestos** como una instrucci√≥n **if**, a trav√©s de conectores l√≥gicos como $or(|)$ y el $and(&)$. Importante que los condicionales se coloquen entre par√©ntesis:

```python
starw['mass'].info()
# mass es object
```

```python
# no toda la serie es float, entonces hacemos:
starw['mass']=starw['mass'].str.replace(',', '').astype(float)# valores con ,
starw.info()
```

Si queremos filtrar los registros para que los personajes seleccionados tengan estatura mayor a 170 cm o con peso superior a los 100 kg:

```python
personajes_pesados_altos= starw[(starw['height']>170) | (starw['mass']>100)]
print('Tama√±o del resultado: ', personajes_pesados_altos.shape)
personajes_pesados_altos
```

# Filas y columnas

el filtrado de filas y columnas al mismo tiempo, puedes obtener tanto los atributos o caracter√≠sticas que te interesan, como los registros que deseas.  

es necesario que utilices los operadores $loc/ilo$c antes de los corchetes. Estos nos permiten que, dentro de los corchetes, especifiquemos las filas que deseamos recuperar y separado con una coma, las columnas que requerimos.

Utiliza el operador **loc** cuando la selecci√≥n sea con base en los datos contenidos; en otras palabras, cuando especifiques nombres de columnas, etiquetas de las filas o expresiones condicionales.  

Por ejemplo, si queremos recuperar √∫nicamente el nombre de los personajes que miden m√°s de 170 cm, realizamos:

```python
personajes_altos=starw.loc[starw["height"]>170,'name']
print('Tama√±o del resultado: ',personajes_altos.shape)
personajes_altos.head()
```

Se pueden recueperar m√°s de una columna:

```python
personajes_ap=starw.loc[starw["height"]>170,['name','height']]
print('Tama√±o del resultado: ',personajes_ap.shape)
personajes_ap.head()
```

El operador **loc** tiene un funcionamiento muy particular:
1. Se establece la etiqueta inicial y a continaici√≥n la etiqueta final
2. Lo que realizar√° **loc** es colocarse en la etiqueta inicial y avanzar, registro a registro (si la segmentaci√≥n es por filas) o columnas ( si la segmentaci√≥n es por columnas), hasta llegar a la etiquieta indicada final.


El operador **iloc** se utiliza cuando la selecci√≥n sea por sub√≠ndices; con base en la posici√≥n del os datos en el dataframe

Por ejemplo, si recuperamos los registros con los sub√≠ndices del 20 al 30 y las columnas del 2 al 6:

```python
personajes_a=starw.iloc[20:30,2:6]
print('Tama√±o del resultado', personajes_a.shape)
personajes_a.head()
```
