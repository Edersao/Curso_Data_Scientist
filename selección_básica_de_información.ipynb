# Filas

En el caso en que desees mantener todas las características o atributos (columnas) de un registro, pero no todos los registros, es decir, **seleccionar filas por condición**.
Para lograrlo, es necesario el uso de condicionales a algunos de los atributos o características de los registros

El filtrado por la condición especificada **incluirá a aquellos registros referidos como filas que la cumplen y excluirá a los que no lo hacen**. Debes colocar el condicional dentro de los corchetes y podrás utilizar los operadores relacionales $<,>,==,!=,>= ,<=$ que ya conoces.
Por ejemplo, podemos obtener todos los personajes de Star Wars que sean más bajos de 100 cm como se muestra en el siguiente script:

```python
personas_bajas=starw[starw['height']<100]
print('Tamaño del producto ', personas_bajas.shape)
personas_bajas
#mantiene todas las columnas pero solo hay 7 filas en el resultado
```

También se puede utilizar **condicionales compuestos** como una instrucción **if**, a través de conectores lógicos como $or(|)$ y el $and(&)$. Importante que los condicionales se coloquen entre paréntesis:

```python
starw['mass'].info()
# mass es object
```

```python
# no toda la serie es float, entonces hacemos:
starw['mass']=starw['mass'].str.replace(',', '').astype(float)# valores con ,
starw.info()
```

Si queremos filtrar los registros para que los personajes seleccionados tengan estatura mayor a 170 cm o con peso superior a los 100 kg:

```python
personajes_pesados_altos= starw[(starw['height']>170) | (starw['mass']>100)]
print('Tamaño del resultado: ', personajes_pesados_altos.shape)
personajes_pesados_altos
```

Para **seleccionar filas por segmentación** se debe colocar entre corchetes $[subindice\_inicial:subíndice\_final]$, con esto recuperas los registros que se encuentren en ese rango. Ejemplo

```python
starw[10:40]
```

# Filas y columnas

el filtrado de filas y columnas al mismo tiempo, puedes obtener tanto los atributos o características que te interesan, como los registros que deseas.  

es necesario que utilices los operadores $loc/ilo$c antes de los corchetes. Estos nos permiten que, dentro de los corchetes, especifiquemos las filas que deseamos recuperar y separado con una coma, las columnas que requerimos.

Utiliza el operador **loc** cuando la selección sea con base en los datos contenidos; en otras palabras, cuando especifiques nombres de columnas, etiquetas de las filas o expresiones condicionales.  

Por ejemplo, si queremos recuperar únicamente el nombre de los personajes que miden más de 170 cm, realizamos:

```python
personajes_altos=starw.loc[starw["height"]>170,'name']
print('Tamaño del resultado: ',personajes_altos.shape)
personajes_altos.head()
```

Se pueden recueperar más de una columna:

```python
personajes_ap=starw.loc[starw["height"]>170,['name','height']]
print('Tamaño del resultado: ',personajes_ap.shape)
personajes_ap.head()
```

El operador **loc** tiene un funcionamiento muy particular:
1. Se establece la etiqueta inicial y a continaición la etiqueta final
2. Lo que realizará **loc** es colocarse en la etiqueta inicial y avanzar, registro a registro (si la segmentación es por filas) o columnas ( si la segmentación es por columnas), hasta llegar a la etiquieta indicada final.


El operador **iloc** se utiliza cuando la selección sea por subíndices; con base en la posición del os datos en el dataframe

Por ejemplo, si recuperamos los registros con los subíndices del 20 al 30 y las columnas del 2 al 6:

```python
personajes_a=starw.iloc[20:30,2:6]
print('Tamaño del resultado', personajes_a.shape)
personajes_a.head()
```
