Documento species.csv de las especies de los personajes de Star Wars.

```python
import pandas as pd
especies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datos/species.csv')
print('Tamaño del archivo:', especies.shape)
especies.head()
```

1. **lower()**: cambia todos los caracteres a minúsculas.
2. **upper**(): cambia todos los caracteres a mayúsculas.
3. **title()**: cambia a mayúsculas todos los inicios de las palabras.
4. **capitaliza()**: Cambia el primer carácter a mayúscula y todos los demás a minúsculas

```python
print('1.',especies['language'].str.lower().head(),sep='\n')
print('2.',especies['language'].str.upper().head(),sep='\n')
print('3.',especies['language'].str.title().head(),sep='\n')
print('4.',especies['language'].str.capitalize().head(),sep='\n')
```

Para conocer la longitud de caracteres de cada "row"

```python
especies['long']=especies['name'].str.len()
especies[['name','long']].head(8)
```

La función <mark>contains()</mark> , permite saber su una alabra o conjunto de caracteres estan dentro de la cadena. Recibe como parámetro la palabra a buscar y regresa $True$ si la encuentra o $False$ si no.
Sensible a mayúsculas y minúsculas, aunque puede ser desabilitada o habilitada usando un segundo parámetro **el atributo case**, que toma valores de $True$ o $False$ respectivamente.
Ej.
Al siguiente dataframe, le aplicamos la búsqueda de la palabra   $blue$ en la columna   $skin_colors$. Como puedes notar solamente regresa una serie con valores booleanos.

```python
especies['skin_colors'].str.contains('blue', case=False).head()
```

Por su parte, la función  <mark>cat()</mark> se utiliza para concatenar cadenas de textos, así podrías combinar el contenido de algunas columnas.

```python
especies['name'].str.cat(' es un '+ especies['classification']).head()
```

Otra forma de usar **cat()** es para generar una sola cadena a partir de todos los valores de una serie, simplemente aplicándola a la columna en cuestión.

```python
especies['name'].str.cat(sep=', ')
```

Para realizar reemplazos de algunos o todos los datos, puedes usar <mark>replace()</mark>. También se puede aplicar en una serie o cadena.

```python
especies['corta_designacion']=especies['designation'].replace({'sentient':'S','reptilian':'R'})
especies[['designation','corta_designacion']].head()
```

En el caso de cadenas:
Ej.
replace() puede aceptar una expresión regular en lugar de una palabra a buscar.

En el suficiente fragmento de script. se ha creado una expresión regular para buscar todas las cadenas que empiezan con (^$T.*)$  . Para todos aquellos registros que cumplan con el patrón, se mandrá la cadena a la función $procesaTexto$. Se especifica el atributo $regex=True$ para indicar que es una verificación a través de una expresión regular.

```python
def procesaTexto(texto):
  return texto.group().upper()
expReg = "(^T.*)"
especies['homeworld change']=especies['homeworld'].str.replace(expReg,procesaTexto, regex=True)
especies['homeworld change'].head(8)
```

método <mark>split()</mark>
la column skin_colors, posee varios colores posibles para la especie. Si quisieramos manupular cada uno de manero individual se aplica un split() para que cada color sea incluido como un lemento de la lista, utilizando un separador ',' :

```python
especies['skin_colors_list']= especies['skin_colors'].str.split(', ')
especies['skin_colors_list'].head()
```

Si quisieramos conocer el tamaño de la lista, se puede hacer con srt para llamar a len()

```python
especies['skin_colors_list_len']= especies['skin_colors'].str.split(', ').str.len()
especies[['skin_colors_list','skin_colors_list_len']].head()
```
