elimina todos los registros que sean de Estados Unidos.

```python
sineu=prueba.loc[prueba['Country']!='United States']
sineu.reset_index(inplace=True)
```

Muestra el número de personas registradas de cada país (ya sin los de Estados Unidos). Se espera una tabla que diga país y número de personas.

```python

person=sineu.groupby(by=['Country'])['Id'].value_counts()
g=person[('Great Britain')].value_counts().iloc[0]
f=person.France.value_counts().iloc[0]
#Actualiza el nombre de los países para que el nombre esté en español.
nperson=pd.DataFrame({'País':['Francia','Gran Bretaña'],'Personas Registradas':[f,g]})
nperson
```
