a. Extraer la informaci√≥n del archivo.

```python
import pandas as pd
vacunas=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datos/1695860027103-country_vaccinations.csv', parse_dates=['date'])
vacunas.head()
```

b. Mostrar la estructura y tipos de datos de cada columna para identificar qu√© operaciones puedes realizar con cada una de ellas, asegur√°ndote que las columnas con fechas sean del tipo **datetime64**.

c. Determinar la cantidad de vacunas aplicadas de cada compa√±√≠a (con base en c√≥mo lo reporta cada pa√≠s en la columna **vaccines**, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
v=vacunas[['country','vaccines','daily_vaccinations','date']]
vp=v.groupby('vaccines').sum()
vp.head()
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo

```python
print('En todo se aplicar√≥n un total de ',vp['daily_vaccinations'].sum(),' vacunas desde ',min(v['date']),' hasta ',max(v['date']))
```

e. Calcular el promedio de vacunas aplicadas por pa√≠s.

```python
vp_m=v.groupby('country').mean().round(2)
print('El promedio de vacunas aplicadas por pais:',vp_m,sep=',')
```

f. Determina la cantidad de vacunas aplicadas el d√≠a 29/01/21 en todo el mundo

```python
vd = vacunas[['date','daily_vaccinations']].groupby([vacunas['date'].dt.year,vacunas['date'].dt.month, vacunas['date'].dt.day]).sum()
print('La cantidad de vacunas aplicadas  el d√≠a 29/01/21 en todo el mundo fue de: ',vd.loc[2021, 1,29][0])
```

g, Crea un dataframe nuevo denominado **conDiferencias** que contenga los datos originales y una columna derivada (_diferencias_) con las diferencias de aplicaci√≥n entre las columnas **daily_vaccinations** y **daiy_vaccinations_raws**.

```python
conDiferencias= vacunas
conDiferencias['diferencias']= conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias.head()
```

h. Obtener el periodo de tiempo entre el registro con fecha m√°s reciente y el registro con fecha m√°s antigua.

```python
print('El periodo de tiempo entre los registros son de ',min(conDiferencias['date']),' hasta ',max(conDiferencias['date']))
print('donde la fecha m√°s reciente es', str(max(conDiferencias['date']))[0:-9],'y la m√°s antigua ',str(min(conDiferencias['date']))[0:-9])
```

i. Crear un dataframe nuevo denominado **conCantidad** que contenga los datos originales y una columna derivada (_canVac_) con la cantidad de vacunas utilizadas cada d√≠a (usar columna vaccines y separar por el car√°cter **,**).

```python
conCantidad=vacunas
conCantidad['canVac']=conCantidad['vaccines'].str.split(', ')
conCantidad.head()
```

j. Generar un dataframe deniminado **antes20**  con todos los registros qe se hayan realizado antes del 20 de diciembre de 2020

```python
antes20=vacunas[vacunas['date']< '2020-12-20']
antes20.head()
```

k. Obtener un dataframe denominado **pfizer** con todos los registros donde se haya utilizado la vacuna _Pfizer_.

```python
pfizer = vacunas[vacunas['vaccines'].str.contains('Pfizer',case=False) == True]
```

l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado **resultadosRetos.xlsx,**, donde cada dataframe ocupe una hoja diferente.

```python
with pd.ExcelWriter("/content/drive/MyDrive/Colab Notebooks/datos/resultadosReto.xlsx") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer,sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="Pfizer")
```
