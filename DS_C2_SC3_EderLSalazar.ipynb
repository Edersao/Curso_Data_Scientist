```python
test=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/manipulación de datos/reto/Test.csv')
test.head(7)
```

```python
test2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/manipulación de datos/reto/Test2.csv')
test2.head(7)
```

```python
train=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/manipulación de datos/reto/Train.csv')
train.head(7)
```

```python
train2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/manipulación de datos/reto/Train2.csv')
train2.head(7)
```

```python
test=test[['Employee_ID', 'Gender', 'Age', 'Education_Level']]
test2=test2[['Employee_ID', ' Computer_OS',' Computer_tickets', ' Mobile_OS',
       ' Mobile_tickets']]
train=train[['Employee_ID', 'Gender', 'Age', 'Education_Level']]
train2=train2[['Employee_ID', ' Computer_OS',' Computer_tickets', ' Mobile_OS',
       ' Mobile_tickets']]
test0=test.merge(test2,left_on='Employee_ID',right_on='Employee_ID',how='inner')
train0=train.merge(train2,left_on='Employee_ID',right_on='Employee_ID',how='inner')
datos=test0.append(train0)
```

5.	Para todos los registros que no tienen un valor para un sistema operativo, ya sea de su computadora o de su dispositivo móvil, asígnales el valor “no registrado”.

```python
d=datos[' Computer_OS'].unique()
datos.loc[~datos[' Computer_OS'].isin( d[:-1]),' Computer_OS']='no registrado'
# ahora moviles
d=datos[' Mobile_OS'].unique()
datos.loc[~datos[' Mobile_OS'].isin( d[:-1]),' Mobile_OS']='no registrado'
datos.tail()
```

6.	Ejecuta la instrucción adecuada en Pandas para encontrar lo siguiente:


  -	El porcentaje de personas que usan Windows, Mac, Linux, no registrado como su sistema operativo en su computadora (se esperan 4 renglones en esta respuesta).



```python
datos['Personas']=1#para facilitar el conteo de personas
porcentaje_sistemaO=datos.groupby([' Computer_OS']).agg({'Personas': np.sum}).apply(lambda x: 100*x/ float(x.sum()))
porcentaje_sistemaO
```

  -	El porcentaje de personas que usan iOS, Android, no registrado como su sistema operativo en su dispositivo móvil (se esperan 3 renglones en esta respuesta).
  


```python
porcentaje_mobile=datos.groupby([' Mobile_OS']).agg({'Personas': np.sum}).apply(lambda x: 100*x/ float(x.sum()))
porcentaje_mobile
```

  -	La edad promedio de la gente que utiliza cada sistema operativo de escritorio.
  


```python
edad_promedio=datos.groupby([' Computer_OS']).agg({'Age': np.mean})
edad_promedio
```


  -	El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio.
  

```python
nivel_educativo=datos.groupby([' Computer_OS']).agg({'Education_Level': np.median})
nivel_educativo
```

  -	El número total de tickets generados por la gente, con cada sistema operativo de escritorio.

```python
total_tickets=datos.groupby([' Computer_OS']).agg({' Computer_tickets': np.sum})
total_tickets
```

7.	Envía el resultado de cada solicitud de información a un archivo .xlsx, donde el resultado de cada consulta esté en su propia hoja de trabajo.

```python
porcentaje_sistemaO.to_excel('/content/drive/MyDrive/Colab Notebooks/manipulación de datos/reto/resultado_reto.xlsx')
with pd.ExcelWriter('/content/drive/MyDrive/Colab Notebooks/manipulación de datos/reto/resultado_reto.xlsx') as writer:
    porcentaje_sistemaO.to_excel(writer, sheet_name='Porcentaje Sistema Operativo')
    porcentaje_mobile.to_excel(writer, sheet_name='Porcentaje Mobile')
    edad_promedio.to_excel(writer,sheet_name='Edad promedio')
    nivel_educativo.to_excel(writer, sheet_name='Nivel educativo')
    total_tickets.to_excel(writer, sheet_name='Tickets totales')
```
